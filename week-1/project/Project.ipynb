```python
#Making a class to hold all values required for the calculations

class interest_plan :
    def __init__(self,principal,rate,time,number):
        self.principal = principal 
        self.rate = rate 
        self.time = time 
        self.number = number

    # Making methods based on the option the user chooses
    # simple interest method
    def simple_interest(self):
        s_i = self.principal * (self.rate / 100) * self.time
        return s_i
    # compound interest method
    def compound_interest(self):
        c_i = self.principal * (1 + ((self.rate/self.number)**self.time))
        return c_i
    # annuity plan method
    def annuity_plan(self):
        a_p = self.principal * ((    (   (1+(self.rate/self.time))  **self.time  ) -1 )   /(self.rate/self.number) )
        return a_p
#Calculation Menu
print("Welcome To the Interest Plan and calculators !\nPlease input which of the following you would like to calculate:")
print("\n||Simple Interest||Compound Interest||Annuity Plan")
while True :

    calc_option = input("\nPlease type the option you want to choose: ")
    print("\nPlease enter your following values:")



    #Sorting the Method by user input
    if calc_option.lower() == "simple interest":
        p = float(input("Principal: "))
        r = float(input("Rate: "))
        t = float(input("time(years): "))
        n = float(input("number of years (n): "))
        simple_I = interest_plan(principal=p,rate=r,time=t,number= n) # puts all user inputs in an object
        finished_calc = simple_I.simple_interest() # runs the simple interest method with user input in an object
        print(f"The simple interest is : {finished_calc}")

    elif calc_option.lower() == "compound interest":
        p = float(input("Principal: "))
        r = float(input("Rate: "))
        t = float(input("time(years): "))
        n = float(input("number of times interest compounds(n): "))
        comp_I = interest_plan(principal=p,rate=r,time=t,number= n) # puts all user inputs in an object
        finished_calc = comp_I.compound_interest() # runs the compound interest method with user input in an object
        print(f"The compound interest is : {finished_calc}")

    elif calc_option.lower() == "annuity plan":
        p = float(input("Principal: "))
        r = float(input("Rate: "))
        t = float(input("time: "))
        n = float(input("number of years times annuity compounds (n): "))
        ann_plan = interest_plan(principal=p,rate=r,time=t,number= n) # puts all user inputs in an object
        finished_calc = ann_plan.annuity_plan() # runs the annuity method with user input in an object
        print(f"The compound interest is : {finished_calc}")
    else :
        print("Failed to select valid options.\n Please retry code")

    end = input("do you want to end the program?(Y/N): ")
    if end.lower() == "y" :
        break
    else:
        continue
```
