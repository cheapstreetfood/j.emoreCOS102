```python
class SST:
    prog1 = "Software Engineering"

    def __init__(self,name):
        self.name = name 
    def speak(self):
        print("My name is",self.name)
        print("I study",SST.prog1)


stud1 = SST(input("Enter student 1 name: "))
stud2 = SST(input("Enter student 2 name: "))

stud1.speak()
stud2.speak()
```

```python
class Addition:
    def __init__(self,f,s):
        self.first = f
        self.second = s
        self.answer = None
    
    def calculate(self):
        self.answer = self.first + self.second
    
    def display(self):
        print("\nFirst: ",self.first)
        print("\nSecond:",self.second)
        print("\Sum: ",self.answer)

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

obj = Addition(num1,num2)
obj.calculate()
obj.display()
```
