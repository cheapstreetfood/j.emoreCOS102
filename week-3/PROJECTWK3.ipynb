```python
import tkinter as tk
import customtkinter as ctk
from tkinter import messagebox as msgbox
import pandas as pd

df = pd.read_csv("giglog.csv",usecols=["Name","Department"])

#The open function for the enter button
def open():
    w_names = df["Name"].tolist() # convert csv column to list
    w_deps = df["Department"].tolist()
    input_name = workername_entry.get() # gets inputs
    input_dep = workerdep_entry.get()
    for name in w_names:
        if name.lower() == input_name.lower(): #checks if any name matches
            x = 1
    for dep in w_deps:
        if dep.lower() == input_dep.lower(): # checks if departments matches
            y = 2
    if x == 1 and y == 2: # checks if both stuff are true
        emp_window() # runs employee window

#Main window
root = tk.Tk()
root.title("GIG Login")
root.geometry("500x500")
root.configure(bg="black")

#Frame for the left side of ui
contact_frame = ctk.CTkFrame(master=root,width=130,border_width=1,border_color="red")
contact_frame.pack(side="left",fill="y",pady=5,padx=5) #puts frame on the left and fills the entire box down

frame_color = contact_frame.cget("fg_color")

contact_email = ctk.CTkButton(master=contact_frame,text="Contact us",
                             text_color="white",font=("Bold",10),
                             width=60,fg_color=frame_color,hover=False)
contact_email.place(x=4,y=56)

contact_phone = ctk.CTkButton(master=contact_frame,text="info@giglogistics.com0",
                             text_color="white",font=("Bold",10),
                             width=60,fg_color=frame_color,hover=False)
contact_phone.place(x=3,y=126)

contact_location = ctk.CTkButton(master=contact_frame,text="Phone:+234813985112",
                             text_color="white",font=("Bold",10),
                             width=60,fg_color=frame_color,hover=False)
contact_location.place(x=2,y=196)

#Name label and box
worker_name = tk.Label(root,text="Name")
worker_name.pack()
workername_entry = tk.Entry(root)
workername_entry.pack()

# Department label and box
worker_dep = tk.Label(root,text="Department")
worker_dep.pack()
workerdep_entry = tk.Entry(root)
workerdep_entry.pack()





# Window for employee
def emp_window():
    emp = tk.Toplevel(root)
    emp.title("Employee window")
    emp.geometry("100x100")


    input_name = workername_entry.get() # gets inputs
    input_dep = workerdep_entry.get()
    w_names = df["Name"].tolist() # convert csv column to list
    w_deps = df["Department"].tolist()

    welcome_label = tk.Label(emp,text=f"Welcome {input_name},from {input_dep} department!\n Others from this department are :")
    welcome_label.pack()

    work_dict = dict(zip(w_names,w_deps)) # Converts the lists I made to dictionaries
    target_names = input_dep #name that matches departments

    matching_names = [key for key,name in work_dict.items() if name == target_names] # makes a list of all names in departments (list comprehension)

    names_label = tk.Label(emp,text=matching_names) 
    names_label.pack()




# The button to enter
enter_button = tk.Button(root,text="Enter",command=open)
enter_button.pack()




root.mainloop()
```
