```python
import numpy as np
#Functions to find roots of polynomials
def quad(a,b,c):
    coefs = [a,b,c]
    roots = np.roots(coefs)
    return roots
def cubc(a,b,c,d):
    coefs = [a,b,c,d]
    roots = np.roots(coefs)
    return roots
def quar(a,b,c,d,e):
    coefs = [a,b,c,d,e]
    roots = np.roots(coefs)
    return roots


print("Welcome to the polynomial root calculator")
print("\n(Quadratic || Cubic ||  Quartic)")
while True: # Makes it a continuous program
    eq = input(print("\nPlease enter the nature or your equation")) # Chooses the equation
    print("\nPlease enter the coefficients of you equation[a,b,c,d,e]")
 
    if eq.lower() == "quadratic": # Matches equation to the function
        a = input("\n a: ") # coeffficients of your polynomials
        b = input("\n b: ")
        c = input("\n c: ")
        root = quad(a,b,c)
        print(f"The roots of this equation are: {root}")
    elif eq.lower() == "cubic":
        a = input("\n a: ") # coeffficients of your polynomials
        b = input("\n b: ")
        c = input("\n c: ")
        d = input("\n d: ")
        root = cubc(a,b,c,d)
        print(f"The roots of this equation are: {root}")
    elif eq.lower() == "quartic":
        a = input("\n a: ") # coeffficients of your polynomials
        b = input("\n b: ")
        c = input("\n c: ")
        d = input("\n d: ")
        e = input("\n e: ")
        root = quar(a,b,c,d,e)
        print(f"The roots of this equation are: {root}")
    else:
        print("Please try again. Select a polynomial from the above")

    choice = input("Would you like to rerun this program?(y/n)")

    if choice.lower() == "n": # Can choose to break from the program
        print("Thank you for using PAU polynomial calculator!")
        break

```
